# Solving by Jax using Gradient Decent Just for Fun

```python
import jax.numpy as jnp
from jax import grad
from jax import vmap
```

```python
import matplotlib.pyplot as plt
```

```python
def cost1(x):
    cost = jnp.array(abs(data - x), float)
    return jnp.array(cost).sum().astype(np.float32)
```

```python
def cost2(x):
    cost = jnp.array((data - x)**2, float)
    return jnp.array(cost).sum().astype(np.float32)
```

# Part 1

```python
gradf = grad(cost1)
loss , values = [] , []
learning_rate , num_epoch, x0 = 0.01, 200 , 100.0
x = jnp.array([x0])
for epoch in range(num_epoch) :
    x = x - learning_rate * gradf(x)
    loss.append(cost1(x))
    values.append(x)
print(int(loss[-1]))
```

```python
color = 'tab:red'
plt.plot(values, color)
plt.ylabel('Alignement position',color=color)
plt.tick_params(axis='y', labelcolor=color)
ax2 = plt.gca().twinx()
color = 'tab:blue'
ax2.plot(loss)
plt.ylabel('Loss', color=color)
ax2.tick_params(axis='y', labelcolor=color)
pass
```
